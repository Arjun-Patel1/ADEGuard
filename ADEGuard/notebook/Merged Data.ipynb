```python
import os
import pandas as pd

# Base directory relative to notebook
base_dir = "../DATA"
years = [2020, 2021, 2022, 2023, 2024, 2025]

# File types and suffixes
file_types = {
    "DATA": "VAERSDATA.csv",
    "SYMPTOMS": "VAERSSYMPTOMS.csv",
    "VAX": "VAERSVAX.csv"
}

# Dictionary to hold merged DataFrames
merged = {key: [] for key in file_types}

# Loop through each year and file type
for year in years:
    folder = f"{year}VAERSData"
    for key, suffix in file_types.items():
        filename = f"{year}{suffix}"
        filepath = os.path.join(base_dir, folder, filename)
        try:
            # Use ISO-8859-1 encoding to avoid UnicodeDecodeError
            df = pd.read_csv(filepath, encoding='ISO-8859-1', low_memory=False)
            df["YEAR"] = year  # Optional: add year column
            merged[key].append(df)
            print(f"✅ Loaded: {filename}")
        except FileNotFoundError:
            print(f"⚠️ Missing file: {filepath}")
        except UnicodeDecodeError as e:
            print(f"❌ Encoding error in {filename}: {e}")

# Output directory for merged files
output_dir = "../merged_output"
os.makedirs(output_dir, exist_ok=True)

# Concatenate and save each merged DataFrame
for key in merged:
    final_df = pd.concat(merged[key], ignore_index=True)
    output_path = os.path.join(output_dir, f"ALL_YEARS_{key}.csv")
    final_df.to_csv(output_path, index=False)
    print(f"📁 Saved: {output_path}")
```

```python
import pandas as pd
import os

# Load merged files
base_path = "../merged_output"
data_df = pd.read_csv(os.path.join(base_path, "ALL_YEARS_DATA.csv"), encoding='ISO-8859-1', low_memory=False)
symptoms_df = pd.read_csv(os.path.join(base_path, "ALL_YEARS_SYMPTOMS.csv"), encoding='ISO-8859-1', low_memory=False)
vax_df = pd.read_csv(os.path.join(base_path, "ALL_YEARS_VAX.csv"), encoding='ISO-8859-1', low_memory=False)

# Select relevant columns (customize as needed)
data_cols = ["VAERS_ID", "AGE_YRS", "SEX", "DIED", "L_THREAT", "DISABLE", "RECOVD", "STATE", "YEAR"]
symptom_cols = ["VAERS_ID", "SYMPTOM1", "SYMPTOM2", "SYMPTOM3"]
vax_cols = ["VAERS_ID", "VAX_TYPE", "VAX_MANU", "VAX_DOSE_SERIES"]

# Filter columns
data_df = data_df[data_cols]
symptoms_df = symptoms_df[symptom_cols]
vax_df = vax_df[vax_cols]

# Merge all on VAERS_ID
merged_df = data_df.merge(symptoms_df, on="VAERS_ID", how="left").merge(vax_df, on="VAERS_ID", how="left")

# Save final CSV
final_path = os.path.join(base_path, "VAERS_MODEL_READY.csv")
merged_df.to_csv(final_path, index=False)
print(f"✅ Final dataset saved to: {final_path}")
```
